# Lesson End Project: Matplotlib

## Problem statement: 

Using an array as input to the dataset, generate a scatterplot, a histogram, subplots, and a line plot.

## Steps to be followed:

1. Importing the necessary libraries 
2. Creating an array using a random number array
3. Generating a scatterplot
4. Generating a histogram
5. Generating subplots
6. Generating a line plot

## Step 1: Importing the Necessary Libraries
- Import numpy and matplotlib

```python
import os
os.environ['MPLBACKEND'] = 'TkAgg'  # Safe backend for most systems
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2: Creating Data Using a Random Number Array
- Use the __random.randn()__ function to generate the array

```python
data = np.random.randn(100)
data
```

## Step 3: Generating the Scatterplot
- Define x-axis as the index of the data
- Define y-axis as the data itself
- Provide labels for the title and x and y axes

```python
plt.figure()
plt.scatter(range(len(data)), data)
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data)
plt.show()
```

## Step 4: Generating a Histogram
- Limit the number of bars to 10
- Provide labels for the title and x and y axes

```python
plt.figure()
plt.hist(data, bins=20)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data)
```

## Step 5: Generating Subplots
- Generate a scatterplot in the first half of the space
- Provide labels for the title, x, and y axes
- Generate a histogram in the second half of the space
- Provide labels for the title and x and y axes

```python
# Subplot
plt.figure()
plt.subplot(2, 1, 1)  
plt.scatter(range(len(data)), data)
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

plt.subplot(2,1,2)
plt.hist(data, bins = 5)
plt.title('Histogram Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

## Step 6: Generating a Line plot
- Use the index of the data in the x-axis
- Use the data on the y axis
- Provide labels for the title and x and y axes

```python
plt.figure()
plt.plot(range(len(data)), data)
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
