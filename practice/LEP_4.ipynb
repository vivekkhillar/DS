## 3. Plotting Various Graphs to Understand Different Columns of the Dataset
- Visualize the age and the number of seconds in the dataset using the subplot method
- Get an overview of the relationship between the columns using the pair plot 

```python
plt.figure()
plt.subplot(2, 1, 1)  
plt.hist(df['seconds'])
plt.title('Histogram')
plt.xlabel('Call duration in seconds')
plt.ylabel('Count')
plt.show()

plt.subplot(2, 1, 2)  
plt.scatter(x='amount', y='seconds', data= df)
plt.title('Scatter plot')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()

```

__Observations:__
- You have got two plots. The first depicts the number of seconds of the calls.
- The second graph shows the price that was charged for the calls, using a scatterplot.

```python
cols =['age','region','calls','sms','data','seconds','amount']
sns.pairplot(df[cols])
plt.show()
```

__Observations__:
- The above graph shows the relationship between various features of the dataset.
- For example, the linear relationship between the number of calls and the amount charged can be clearly made out. 

## Step 4: Detecting Outliers Using the Boxplot
- Boxplot helps to visualize the outliers when 2 columns are put in x and y axes

```python
plt.subplot(2,1,1)
sns.boxplot(x= 'region', y='calls', data=df)
plt.show()

plt.subplot(2,1,2)
sns.boxplot(x ='network', y= 'calls', data=df)
plt.show()
```

__Observations:__
- The outliers for the feature calls when compared with each network and region is illustrated above.

### 5. Visualizing the Correlation Between the Various Columns of the Dataset
- Correlation can be visualized using the __corr()__ method

```python
sns.heatmap(df.corr())
plt.show()

# winsorization is removes the outliers by replacing them with the nearest non-outlier value
```
